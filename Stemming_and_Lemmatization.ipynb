## Stemming and Lemmatization

### Stemming
**What it is:** Stemming is like chopping off the ends of words to find their root form. It's a quick and dirty way to reduce words to their base form.

**Example:**
- "running" $\rightarrow$ "run"
- "studies" $\rightarrow$ "studi"
- "cats" $\rightarrow$ "cat"

### Lemmatization
**What it is:** Lemmatization is more sophisticated. It uses a dictionary to find the base form of a word, considering its grammatical context. It aims to produce actual words, not just chopped-off pieces.

**Example:**
- "running" $\rightarrow$ "run" (correct base form)
- "studies" $\rightarrow$ "study" (correct base form)
- "better" $\rightarrow$ "good" (base form considering grammatical context)

## Key Differences
**Accuracy:** Lemmatization is generally more accurate than stemming because it uses a dictionary and considers grammatical context.

**Complexity:** Lemmatization is more computationally expensive than stemming.

**Use Cases:**

**Stemming:** Useful for tasks like search engines, where speed is important.

**Lemmatization:** Better for tasks like natural language processing (NLP), where accuracy is crucial.

### Simple Analogy
Imagine you have a bunch of toys, and you want to group them by type.

**Stemming:** You might just look at the beginning of the word and put all toys that start with "car" together (car, cars, cartoon, etc.).

**Lemmatization:** You'd look up each toy in a dictionary and find its true "type" (car, car, cartoon, etc.).

### In Summary:
**Stemming** is a fast and simple way to reduce words to their root form, but it can sometimes create non-words.

**Lemmatization** is more accurate but slower, producing actual words and considering grammatical context.

```python
# import Stemmer and tokenize library
import nltk
from nltk.stem import PorterStemmer
from nltk.tokenize import word_tokenize
```

```python
# download 'punkt' resources
import nltk
nltk.download('punkt')
nltk.download('punkt_tab')
```

```python
# run the code
porter_stem = PorterStemmer()

# creating an instance of the code
text = "we are all learning how to train our generative model which help us hit the ground running on our professional carrer while making cool stuff."
tokens = word_tokenize(text)

print(tokens)
```

```python
# split the input in token to the process them further
for token in tokens:
    # print(token, " : ", porter_stem.stem(token),end=", ")
    print(token, " : ", porter_stem.stem(token))
```

```python
stemmed_word = [porter_stem.stem(token) for token in tokens]
print(stemmed_word)

# iterate through the token and perform stemming on each token and perform stemming on each token

print(text, "\n")
# join
print(" ".join(stemmed_word))
```

# Lemmatization Demo

```python
# import lammetization and tokenize library
import nltk
from nltk.stem import WordNetLemmatizer
from nltk.tokenize import word_tokenize
from nltk.corpus import wordnet
```

```python
# Creating the instance
lem = WordNetLemmatizer()

text = "we are all learning how to train our generative model which help us hit the ground running on our professional carrer while making cool stuff."
tokens = word_tokenize(text)

tokens
```

```python
# function nltk_tag_wordnet_tag for adjective, verb, noun, adv
def nltk_tag_to_wordnet_tag(nltk_tag):
    if nltk_tag.startswith('J'):
        return wordnet.ADJ
    elif nltk_tag.startswith('V'):
        return wordnet.VERB
    elif nltk_tag.startswith('N'):
        return wordnet.NOUN
    elif nltk_tag.startswith('R'):
        return wordnet.ADV
    else:
        return None
```

```python
import nltk
nltk.download('averaged_perceptron_tagger')
```

```python
import nltk

# Naya updated resource download karein
nltk.download('averaged_perceptron_tagger_eng')

```

```python
# Phir se preview chalayein
pos_tagged_tokens = nltk.pos_tag(tokens)
pos_tagged_tokens
```

```python
# 1. Khali list banayein result save karne ke liye
lematized_word = []

# 2. WordNet resources download karein (agar pehle nahi kiye)
import nltk
nltk.download('wordnet')
```

```python

# 3. Loop chalayein jo har word aur uske tag ko process kare
for token, tag in pos_tagged_tokens:
    # Helper function se WordNet wala tag hasil karein
    wordnet_tag = nltk_tag_to_wordnet_tag(tag)

    if wordnet_tag is None:
        # Agar tag nahi mila toh simple lemmatize karein
        lematized_word.append(token)
    else:
        # Agar tag mil gaya (Verb, Noun etc) toh uske sath lemmatize karein
        lematized_word.append(lem.lemmatize(token, wordnet_tag))

# Result print karein
print(lematized_word)

# 4. Join karke poora sentence banayein
print("\nOriginal Text:")
print(text)

print("\nLemmatized Text:")
print(" ".join(lematized_word))
```
